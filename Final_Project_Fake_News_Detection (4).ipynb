```python
def clean_tweet(text):

    # Remove URLs
    text = re.sub(r'http\S+|www.\S+', '', text)

    # Remove usernames (e.g., @user)
    text = re.sub(r'@\w+', '', text)

    # Remove hashtags (but keep the word)
    text = re.sub(r'#', '', text)

    # Remove numbers
    text = re.sub(r'\d+', '', text)

    # Remove punctuation and special characters
    text = re.sub(r'[^\w\s]', '', text)

    # Remove extra spaces
    text = re.sub(r'\s+', ' ', text).strip()

    return text
```

```python
df['text'] = df['text'].apply(clean_tweet)
```

# **Model Selection and Integration**

Load the pre-trained BERT or DistilBERT model and tokenizer from Hugging Face.


```python
from transformers import BertModel

# Load pre-trained BERT model (base model, not classifier)
model = BertModel.from_pretrained('bert-base-uncased')
```

```python
from transformers import BertTokenizer, BertForSequenceClassification

tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
model = BertForSequenceClassification.from_pretrained('bert-base-uncased', num_labels=2)
```

Preprocess the text data to create input embeddings compatible with the selected model.


```python
train_encodings = tokenizer(
    train_df["text"].tolist(),
    truncation=True,
    padding=True,
    max_length=512,
    return_tensors="pt"
)
```

```python
import os
os.environ["WANDB_DISABLED"] = "true"
```

```python
from transformers import Trainer, TrainingArguments
from torch.utils.data import Dataset
import torch

# Dataset class
class NewsDataset(Dataset):
    def __init__(self, encodings, labels):
        self.encodings = encodings
        self.labels = labels
    def __getitem__(self, idx):
        item = {key: val[idx] for key, val in self.encodings.items()}
        item['labels'] = torch.tensor(self.labels[idx])
        return item
    def __len__(self):
        return len(self.labels)

# Convert labels
train_labels = train_df["lable"].tolist()
val_labels = val_df["lable"].tolist()

# Create datasets
train_dataset = NewsDataset(train_encodings, train_labels)
val_encodings = tokenizer(val_df["text"].tolist(), truncation=True, padding=True, max_length=512, return_tensors="pt")
val_dataset = NewsDataset(val_encodings, val_labels)

# Training arguments
training_args = TrainingArguments(
    output_dir='./results',
    num_train_epochs=3,
    per_device_train_batch_size=8,
    per_device_eval_batch_size=8,
    eval_strategy="epoch",
    logging_dir='./logs',
    logging_steps=10
)

# Trainer
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=train_dataset,
    eval_dataset=val_dataset
)

# Train
trainer.train()
```

```python
# Prepare test encodings and dataset
test_labels = test_df["lable"].tolist()
test_encodings = tokenizer(
    test_df["text"].tolist(),
    truncation=True,
    padding=True,
    max_length=512,
    return_tensors="pt"
)
test_dataset = NewsDataset(test_encodings, test_labels)

# Predict
from sklearn.metrics import accuracy_score

predictions = trainer.predict(test_dataset)
pred_labels = predictions.predictions.argmax(axis=1)

# Accuracy
accuracy = accuracy_score(test_labels, pred_labels)
print("Test Accuracy:", accuracy)
```

```python
import streamlit as st
from transformers import BertTokenizer, BertForSequenceClassification
import torch
import torch.nn.functional as F

# Load model and tokenizer

from transformers import BertTokenizer, BertForSequenceClassification

model = BertForSequenceClassification.from_pretrained(
    "/content/drive/MyDrive/saved_model",
    local_files_only=True
)
tokenizer = BertTokenizer.from_pretrained(
    "/content/drive/MyDrive/saved_model",
    local_files_only=True
)


# UI
st.title("Fake News Detection Chatbot")
st.write("Enter a news snippet below to check if it's real or fake:")

user_input = st.text_area("Paste a news article or snippet here:")

# Button logic
if st.button("Check News"):
    if user_input.strip() == "":
        st.warning("Please enter some text.")
    else:
        inputs = tokenizer(user_input, return_tensors="pt", truncation=True, padding=True, max_length=512)
        with torch.no_grad():
            outputs = model(**inputs)
            probs = F.softmax(outputs.logits, dim=1)
            confidence, pred = torch.max(probs, dim=1)

        # Set label and color based on prediction
        if pred.item() == 1:
            label = "True"
            color = f"rgba(0, 128, 0, {confidence.item()})"  # Green
        else:
            label = "Fake"
            color = f"rgba(220, 20, 60, {confidence.item()})"  # Red

        # Display result with dynamic background
        st.markdown(
            f"<div style='background-color:{color}; padding:10px; border-radius:10px; color:white; font-weight:bold;'>"
            f"This news is likely <strong>{label}</strong> with a confidence of {confidence.item() * 100:.2f}%"
            f"</div>",
            unsafe_allow_html=True
        )
```
